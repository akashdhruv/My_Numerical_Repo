```python
import numpy
import matplotlib.pyplot as plt
from math import *
%matplotlib inline


# Constants
g = 9.8      
v_t = 4.9    
C_D = 1/5.  
C_L = 1.0    
v0 = 4.9
x0 = 0.0 



# function f to store the time derivates
def f(u):
    v = u[0]
    theta = u[1]
    x = u[2]
    y = u[3]
    
    return numpy.array([-g*sin(theta) - C_D/C_L*g/v_t**2*v**2,
                      -g*cos(theta)/v + g/v_t**2*v,
                      v*cos(theta),
                      v*sin(theta)])

# function to solve euler equations
def euler_step(u, f, dt):
    return u + dt * f(u)

def rk_2(u, f, dt):
    ust=u + 0.5*dt * f(u)
    return u+dt*f(ust)

# function to get airplane path
def get_airplane_path(y0,theta0):
    T = 10.0                         
    dt = 0.001 
    N = int(T/dt) + 1 
    t = numpy.linspace(0.0, T, N)
    u = numpy.empty((N, 4))
    u[0] = numpy.array([v0, theta0, x0, y0])
    for n in range(N-1):    
        u[n+1] = rk_2(u[n], f, dt)
        if(u[n+1,3]<0):
            break
    return u,n


# Optimisation Algorithm

# Specifying the population arrays of intial conditions
Np=5
D=2
genmax=10
F=0.5
CR=0.9

a=np.zeros((Np,D),dtype=float)
b=np.zeros((Np,D),dtype=float)
diff=np.zeros(genmax,dtype=float)



# lower bounds of initial conditions
for i in range(Np):
    a[i][0]=2.0
    a[i][1]=-pi/4

# upper bounds of initial conditions
for i in range(Np):
    b[i][0]=2.413
    b[i][1]=pi/4


x=a+(b-a)*(np.random.random((Np,D)))

print "Initial population:"
print x
```
