```python
import numpy as np
import matplotlib.pyplot as plt


# Constants
g = 9.8       
rho=1.019
r=0.5
ve=325.
C_d=0.15
mf=100.0
ms=50.0
mdot=20.0
v0=0.0
h0=0.0

# function f to store the time derivates
def f(u):
    v = u[0]
    h = u[1]
    return numpy.array([-g + (mdot/(ms+mf))*ve- ((1./2.)*rho*v*abs(v)*numpy.pi*r**2*C_d)/(ms + mf), v])

# function to solve euler equations
def euler_step(u, f, dt):
    return u + dt * f(u)



def get_rocket_data(v0,h0):
    T = 40.0                         
    dt = 0.1 
    N = int(T/dt) + 1 
    t = numpy.linspace(0.0, T, N)
    u = numpy.empty((N, 2))
    u[0] = numpy.array([v0, h0])
    global mf
    global mdot
    for n in range(N-1):
        mf=mf-mdot*dt
        if (mf<=0):
            mf=0
            mdot=0
        u[n+1] = euler_step(u[n], f, dt)
        print n, u[n+1,1]
        if(u[n+1,1]<0):
            break
    return u,t,n


[u,t,n]=get_rocket_data(v0,h0)

v=u[0:n,0]
h=u[0:n,1]
t=t[0:n]

plt.figure
plt.plot(t,v)
plt.show()

plt.figure
plt.plot(t,h)
plt.show()

(vmax,index) = max((v,i) for i,v in enumerate(v))

print "Maximum Velocity: ",vmax,"m/s"
print "Time and Height at this point: ",t[index],"s, ",h[index],"m"

(hmax,index) = max((h,i) for i,h in enumerate(h))

print "Maximum Height: ",hmax,"m"
print "Time and Velocity at this point: ",t[index],"s, ",v[index],"m/s"


print t[-1],v[-1]
```
